```python
#loading the libraries

import pandas as pd
import numpy as np
```

```python
data_path = '/Users/mscorzoni/Downloads/Data Collection/quant_alunos_2023_2.csv'

data =  pd.read_csv(data_path)
data.head()
```

```python
# calculate 'RAZAO_CLASSE_ESP_ENSINO'
data['RAZAO_CLASSE_ESP_ENSINO'] = data['CLASSES ESPECIAIS'] / (data['ANOS INICIAIS'] + data['ANOS FINAIS'])
```

```python
# substitute nan 
data['RAZAO_CLASSE_ESP_ENSINO'] = data['RAZAO_CLASSE_ESP_ENSINO'].where((data['CLASSES ESPECIAIS'] != 0) & (data['ANOS INICIAIS'] + data['ANOS FINAIS'] != 0))
```

```python
# discretization using qcut

data['RAZAO_CLASSE_ESP_ENSINO_Q'] = pd.qcut(data['RAZAO_CLASSE_ESP_ENSINO'], 5, duplicates='drop')

data.head()
```

```python
# discretization cut
max_value =  data['RAZAO_CLASSE_ESP_ENSINO'].max()
bins = [0, 0.025, 0.05, 0.075, 0.1, max_value]
data['RAZAO_CLASSE_ESP_ENSINO_I'] = pd.cut(data['RAZAO_CLASSE_ESP_ENSINO'], bins=bins, right=False)
```

```python
# counting data last intervals to variables discretizables

q_count = data['RAZAO_CLASSE_ESP_ENSINO_Q'].value_counts().iloc[-1]
i_count = data['RAZAO_CLASSE_ESP_ENSINO_I'].value_counts().iloc[-1]

q_count, i_count
```
