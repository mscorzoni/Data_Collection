```python
#loading the libraries

import pandas as pd
import numpy as np
```

```python
data_path = '/Users/mscorzoni/Downloads/Data Collection/quant_alunos_2023_2.csv'

data =  pd.read_csv(data_path)
data.head()
```

```python
# calculate 'RAZAO_CLASSE_ESP_ENSINO'
data['RAZAO_CLASSE_ESP_ENSINO'] = data['CLASSES ESPECIAIS'] / (data['ANOS INICIAIS'] + data['ANOS FINAIS'])
```

```python
# substitute nan 
data['RAZAO_CLASSE_ESP_ENSINO'] = data['RAZAO_CLASSE_ESP_ENSINO'].where((data['CLASSES ESPECIAIS'] != 0) & (data['ANOS INICIAIS'] + data['ANOS FINAIS'] != 0))
```

```python
# discretization using qcut

data['RAZAO_CLASSE_ESP_ENSINO_Q'] = pd.qcut(data['RAZAO_CLASSE_ESP_ENSINO'], 5, duplicates='drop')

data.head()
```

```python
# discretization cut
max_value =  data['RAZAO_CLASSE_ESP_ENSINO'].max()
bins = [0, 0.025, 0.05, 0.075, 0.1, max_value]
data['RAZAO_CLASSE_ESP_ENSINO_I'] = pd.cut(data['RAZAO_CLASSE_ESP_ENSINO'], bins=bins, right=False)
```

```python
# counting data last intervals to variables discretizables

q_count = data['RAZAO_CLASSE_ESP_ENSINO_Q'].value_counts().iloc[-1]
i_count = data['RAZAO_CLASSE_ESP_ENSINO_I'].value_counts().iloc[-1]

q_count, i_count
```

```python
# importing the correct packages

from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler

```

```python
# z-score calculation

data['ENSINO MEDIO_z'] = zscore(data['ENSINO MEDIO'])
```

```python
# setting up the MinMaxScaller for normalization to the range 0 to 1000

scaler = MinMaxScaler(feature_range=(0,1000))
data['ENSINO MEDIO_mm'] = scaler.fit_transform(data[['ENSINO MEDIO']])
data.head()
```

```python
# calculating metrics

mean_z = data['ENSINO MEDIO_z'].mean()
median_z = data['ENSINO MEDIO_z'].median()
mean_mm = data['ENSINO MEDIO_mm'].mean()
median_mm = data['ENSINO MEDIO_mm'].median()
```

```python
round(mean_z, 1), round(median_z, 1), round(mean_mm, 1), round(median_mm, 1)
```

```python
# removing nan

cols_to_check = ['EJA FUNDAMENTAL_AI', 'EJA ENSINO MÉDIO', 'EJA FUNDAMENTAL_AF', 'EJA FUNDAMENTAL_MULTI']
data_clean =  data.dropna(subset=cols_to_check)
data_clean.head()
```

```python
# creating new attributes
data_clean['EJA_TOTAL'] = data_clean['EJA FUNDAMENTAL_AI'] + data_clean['EJA ENSINO MÉDIO'] + data_clean['EJA FUNDAMENTAL_AF'] + data_clean['EJA FUNDAMENTAL_MULTI']
data_clean['FUND_MEDIO_TOTAL'] = data_clean['ANOS INICIAIS'] + data_clean['ANOS FINAIS'] + data_clean['ENSINO MEDIO']

data_clean.head()


```

```python
# log transformation

data_clean['EJA_TOTAL_L'] = np.log2(data_clean['EJA_TOTAL']+1)
data_clean['FUND_MEDIO_TOTAL_L'] = np.log2(data_clean['FUND_MEDIO_TOTAL']+1)
```

```python
# correlaction matrix

correlation_matrix = data_clean[['EJA_TOTAL', 'FUND_MEDIO_TOTAL', 'EJA_TOTAL_L', 'FUND_MEDIO_TOTAL_L']].corr()
correlation_matrix
```
